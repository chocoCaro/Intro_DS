### **IMPORT PACKAGES**

```python
import pandas as pd
import numpy as np
```

### **LOAD DATASETS**

```python
# nba.com
nba_achievements = pd.read_csv(filepath_or_buffer='data_[nba.com]/nba_achievements.csv')
nba_all_time_records = pd.read_csv(filepath_or_buffer='data_[nba.com]/nba_all_time_records.csv')
nba_hall_of_fame = pd.read_csv(filepath_or_buffer='data_[nba.com]/nba_hall_of_fame.csv')
nba_roster = pd.read_csv(filepath_or_buffer='data_[nba.com]/nba_roster.csv')
nba_teams = pd.read_csv(filepath_or_buffer='data_[nba.com]/nba_teams.csv')

# basketball-reference.com
nba_active_teams = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/nba_active_teams.csv')
nba_defunct_teams = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/nba_defunct_teams.csv')
nba_players = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/nba_players.csv')
season_meta = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/season_meta.csv')
team_stats = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/team_stats.csv')
team_stats_per_game = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/team_stats_per_game.csv')
player_stats = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/player_stats.csv')
player_stats_per_game = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/player_stats_per_game.csv')
team_ratings = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/team_ratings.csv')
rookie_stats = pd.read_csv(filepath_or_buffer='data_[basketball-reference.com]/rookie_stats.csv')
```

### **nba_achievements.csv**

```python
nba_achievements.columns = nba_achievements.columns.str.lower().str.replace(' ', '_')
nba_achievements.rename(columns={'team_name': 'team'}, inplace=True)
nba_achievements['team'] = nba_achievements['team'].str.title()
nba_achievements
```

### **nba_all_time_records.csv**

```python
nba_all_time_records.drop(columns='Player Link', inplace=True)
nba_all_time_records.columns = nba_all_time_records.columns.str.lower().str.replace(' ', '_')
nba_all_time_records.rename(columns={'team_name': 'team'}, inplace=True)
nba_all_time_records['team'] = nba_all_time_records['team'].str.title()
nba_all_time_records
```

### **nba_hall_of_fame.csv**

```python
nba_hall_of_fame.drop(columns='Player Link', inplace=True)
nba_hall_of_fame.columns = nba_hall_of_fame.columns.str.lower().str.replace(' ', '_')
nba_hall_of_fame.rename(columns={'team_name': 'team', 'seasons_with_team': 'seasons'}, inplace=True)
nba_hall_of_fame['team'] = nba_hall_of_fame['team'].str.title()
nba_hall_of_fame
```

### **nba_roster.csv**

```python
nba_roster.drop(columns=['Birthdate', 'School', 'Player Link'], inplace=True)
nba_roster.columns = nba_roster.columns.str.lower().str.replace(' ', '_')
nba_roster.rename(columns={'team_name': 'team'}, inplace=True)
nba_roster['team'] = nba_roster['team'].str.title()
nba_roster['#'] = nba_roster['#'].str.replace('#', '')
nba_roster['height'] = nba_roster['height'].str.split('-').apply(lambda x: int(x[0]) * 12 + int(x[1]))
nba_roster['weight'] = nba_roster['weight'].str.split().str[0]
nba_roster['season'] = nba_roster['season'].str[:5] + (nba_roster['season'].str[:4].astype(int) + 1).astype(str)
nba_roster
```

### **nba_teams.csv & nba_active_teams**

```python
nba_teams = nba_teams.loc[:, ['Division', 'Team Name']]
nba_teams['Division'] = nba_teams['Division'].str.title()
nba_teams.rename(columns={'Division': 'division', 'Team Name': 'team'}, inplace=True)
nba_teams.head()
```

```python
nba_active_teams = nba_active_teams[nba_active_teams['end_year'] == '2024-25'].drop_duplicates(subset='franchise', ignore_index=True)
nba_active_teams.drop(columns='end_year', inplace=True)
nba_active_teams.rename(columns={'franchise': 'team', 'start_year': 'start_season'}, inplace=True)
nba_active_teams['start_season'] = nba_active_teams['start_season'].str[:5] + (nba_active_teams['start_season'].str[:4].astype(int) + 1).astype(str)
nba_active_teams.head()
```

```python
nba_active_teams = nba_teams.merge(right=nba_active_teams, on='team')
nba_active_teams.head()
```

### **nba_defunct_teams.csv**

```python
nba_defunct_teams.rename(columns={'franchise': 'team', 'start_year': 'start_season', 'end_year': 'end_season'}, inplace=True)
nba_defunct_teams['start_season'] = nba_defunct_teams['start_season'].str[:5] + (nba_defunct_teams['start_season'].str[:4].astype(int) + 1).astype(str)
nba_defunct_teams['end_season'] = nba_defunct_teams['end_season'].str[:5] + (nba_defunct_teams['end_season'].str[:4].astype(int) + 1).astype(str)
nba_defunct_teams.head()
```

### **nba_players.csv**

```python
nba_players.columns = nba_players.columns.str.lower().str.replace(' ', '_')
nba_players['height'] = nba_players['height'].str.split('-').apply(lambda x: int(x[0]) * 12 + int(x[1]))
nba_players['birth_date'] = pd.to_datetime(arg=nba_players['birth_date'], format='%B %d, %Y')
nba_players
```

### **season_meta.csv**

```python
season_meta.columns = season_meta.columns.str.lower().str.replace(' ', '_')
season_meta.head()
```

### **team_stats.csv**

```python
team_stats.columns = team_stats.columns.str.lower().str.replace(' ', '_')
team_stats.rename(columns={'rk': 'rank'}, inplace=True)
team_stats
```

### **team_stats_per_game.csv**

```python
team_stats_per_game.columns = team_stats_per_game.columns.str.lower().str.replace(' ', '_')
team_stats_per_game.rename(columns={'rk': 'rank'}, inplace=True)
team_stats_per_game
```

### **player_stats.csv**

```python
team_mapping = {
    "MIL": "Milwaukee Bucks", 
    "DEN": "Denver Nuggets", 
    "OKC": "Oklahoma City Thunder", 
    "CHA": "Charlotte Hornets", 
    "DAL": "Dallas Mavericks", 
    "BOS": "Boston Celtics", 
    "LAL": "Los Angeles Lakers", 
    "MIN": "Minnesota Timberwolves", 
    "SAC": "Sacramento Kings", 
    "PHO": "Phoenix Suns", 
    "PHI": "Philadelphia 76ers", 
    "NYK": "New York Knicks", 
    "ORL": "Orlando Magic", 
    "MIA": "Miami Heat", 
    "DET": "Detroit Pistons", 
    "SAS": "San Antonio Spurs", 
    "TOR": "Toronto Raptors", 
    "LAC": "La Clippers", 
    "CLE": "Cleveland Cavaliers", 
    "GSW": "Golden State Warriors", 
    "CHI": "Chicago Bulls", 
    "MEM": "Memphis Grizzlies", 
    "ATL": "Atlanta Hawks", 
    "WAS": "Washington Wizards", 
    "IND": "Indiana Pacers", 
    "BKN": "Brooklyn Nets", 
    "HOU": "Houston Rockets", 
    "UTA": "Utah Jazz", 
    "POR": "Portland Trail Blazers", 
    "NOP": "New Orleans Pelicans",
    "NJN": "New Jersey Nets",
    "BRK": "Brooklyn Nets",
    "SEA": "Seattle SuperSonics",
    "CHO": "Charlotte Hornets",
    "2TM": "Total",
    "3TM": "Total",
}
```

```python
player_stats.columns = player_stats.columns.str.lower().str.replace(' ', '_')
player_stats['team'] = np.where(player_stats['team'].isin(['NOH', 'NOK']), 'NOP', player_stats['team'])
player_stats['team'] = player_stats['team'].map(team_mapping)
player_stats
```

```python
player_stats[player_stats['season'].isna()]
```

```python
player_stats.dropna(subset='season', inplace=True)
player_stats['season'] = player_stats['season'].str[:5] + (player_stats['season'].str[:4].astype(int) + 1).astype(str)
player_stats.head()
```

### **player_stats_per_game.csv**

```python
player_stats_per_game.columns = player_stats_per_game.columns.str.lower().str.replace(' ', '_')
player_stats_per_game['team'] = np.where(player_stats_per_game['team'].isin(['NOH', 'NOK']), 'NOP', player_stats_per_game['team'])
player_stats_per_game['team'] = player_stats_per_game['team'].map(team_mapping)
player_stats_per_game
```

```python
player_stats_per_game[player_stats_per_game['season'].isna()]
```

```python
player_stats_per_game.dropna(subset='season', inplace=True)
player_stats_per_game.head()
```

### **team_ratings.csv**

```python
team_ratings.columns = team_ratings.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
team_ratings
```

### **rookie_stats.csv**

```python
rookie_stats.columns = rookie_stats.columns.str.lower().str.replace(' ', '_')
rookie_stats['debut_date'] = rookie_stats['debut'].str[:11].str.replace("'", '')
rookie_stats['debut_date'] = pd.to_datetime(arg=rookie_stats['debut_date'], format='mixed')
rookie_stats['debut_match'] = rookie_stats['debut'].str[12:]
rookie_stats.drop(columns='debut', inplace=True)
rookie_stats
```

### **Write cleaned data to file csv**

```python
nba_achievements.to_csv(path_or_buf='final_data/nba_achievements.csv', index=False)
nba_all_time_records.to_csv(path_or_buf='final_data/nba_all_time_records.csv', index=False)
nba_hall_of_fame.to_csv(path_or_buf='final_data/nba_hall_of_fame.csv', index=False)
nba_roster.to_csv(path_or_buf='final_data/nba_roster.csv', index=False)
nba_active_teams.to_csv(path_or_buf='final_data/nba_active_teams.csv', index=False)
nba_defunct_teams.to_csv(path_or_buf='final_data/nba_defunct_teams.csv', index=False)
nba_players.to_csv(path_or_buf='final_data/nba_players.csv', index=False)
season_meta.to_csv(path_or_buf='final_data/season_meta.csv', index=False)
team_stats.to_csv(path_or_buf='final_data/team_stats.csv', index=False)
team_stats_per_game.to_csv(path_or_buf='final_data/team_stats_per_game.csv', index=False)
player_stats.to_csv(path_or_buf='final_data/player_stats.csv', index=False)
player_stats_per_game.to_csv(path_or_buf='final_data/player_stats_per_game.csv', index=False)
team_ratings.to_csv(path_or_buf='final_data/team_ratings.csv', index=False)
rookie_stats.to_csv(path_or_buf='final_data/rookie_stats.csv', index=False)
```
