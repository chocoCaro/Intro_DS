### **LOAD DATASETS**

```python
nba_teams = pd.read_csv(filepath_or_buffer='Data_NBA_1/nba_teams.csv')
nba_roster = pd.read_csv(filepath_or_buffer='Data_NBA_1/nba_roster.csv')
nba_team_stats = pd.read_csv(filepath_or_buffer='Data_NBA_1/nba_team_stats.csv')
nba_stats = pd.read_csv(filepath_or_buffer='Data_NBA_1/nba_stats.csv')
nba_hall_of_fame = pd.read_csv(filepath_or_buffer='Data_NBA_1/nba_hall_of_fame.csv')

nba_active_teams = pd.read_csv(filepath_or_buffer='Data_NBA_2/nba_active_teams.csv')
nba_defunct_teams = pd.read_csv(filepath_or_buffer='Data_NBA_2/nba_defunct_teams.csv')
nba_players = pd.read_csv(filepath_or_buffer='Data_NBA_2/nba_players.csv')
team_stats = pd.read_csv(filepath_or_buffer='Data_NBA_2/team_stats.csv')
player_stats = pd.read_csv(filepath_or_buffer='Data_NBA_2/player_stats.csv')
```

```python
nba_teams = nba_teams.loc[:, ['Division', 'Team Name']]
nba_teams['Division'] = nba_teams['Division'].str.title()
nba_teams.rename(columns={'Division': 'division', 'Team Name': 'franchise'}, inplace=True)
nba_teams.head()
```

```python
nba_active_teams = nba_active_teams[nba_active_teams['end_year'] == '2024-25'].drop_duplicates(subset='franchise', ignore_index=True)
nba_active_teams.drop(columns=['league', 'end_year'], inplace=True)
nba_active_teams.head()
```

```python
nba_active_teams = nba_teams.merge(right=nba_active_teams, on='franchise')
nba_active_teams.head()
```
